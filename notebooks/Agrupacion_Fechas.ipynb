## Dias Festivos a Obtener

```python
import datetime
```

```python
def leap_year(year: str) -> bool:
  year = int(year)
  return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)


def obtain_days(month: str, year: str):
    # Abril, junio, septiembre y noviembre tienen 30
    if month in ['4', '6', '9', '11']:
        return 30
    # Febrero depende de si es o no bisiesto
    if month == '2':
        if leap_year(year):
            return 29
        else:
            return 28
    else:
        # En caso contrario, tiene 31 días
        return 31
```

```python
def get_new_holidays(date, diff):
  if diff == 'one':
    days_month = obtain_days(str(date.month), str(date.year))
 
    month_str = ''
    year_str = ''
    day_str = ''

    day= int(date.day) +1
    if day > days_month:
      day_str = '01'
      if date.month+1 > 12:
        year_str = str(date.year+1)
        month_str = '01'
      else:
        month_str = str(date.month+1)
        year_str = str(date.year)
    else:
      day_str =str(date.day+1)
      year_str = str(date.year)
      month_str = str(date.month)

    date_str = year_str+'-'+month_str+'-'+day_str
    return datetime.datetime.strptime(date_str, '%Y-%m-%d')

  if diff == 'two':
    days_month = obtain_days(str(date.month), str(date.year))
 
    month_str1 = ''
    year_str1 = ''
    day_str1 = ''

    day=int(date.day) +1
    if day > days_month:
      day_str1 = '01'
      if date.month+1 > 12:
        year_str1 = str(date.year+1)
        month_str1 = '01'
      else:
        month_str1 = str(date.month+1)
        year_str1 = str(date.year)
    else:
      day_str1 =str(date.day+1)
      year_str1 = str(date.year)
      month_str1 = str(date.month)

    date_str1 = year_str1+'-'+month_str1+'-'+day_str1
 
    month_str2 = ''
    year_str2 = ''
    day_str2 = ''

    day = int(date.day)+2
    if day > days_month:
      day_str2 = '01'
      if date.month+1 > 12:
        year_str2 = str(date.year+1)
        month_str2 = '01'
      else:
        month_str2 = str(date.month+1)
        year_str2 = str(date.year)
    else:
      day_str2 =str(date.day+2)
      year_str2 = str(date.year)
      month_str2 = str(date.month)

    date_str2 = year_str2+'-'+month_str2+'-'+day_str2

    return datetime.datetime.strptime(date_str1, '%Y-%m-%d') ,datetime.datetime.strptime(date_str2, '%Y-%m-%d')
```

```python

def to_date(date):
  return datetime.datetime.strptime(date, '%Y-%m-%d')
  
def to_string(date):
  date_ret = date.strftime('%Y-%m-%d')
  return date_ret

def create_new_row(date, df):
  municipios = df['Municipio'].tolist()
  municipios=set(municipios)
  month = df['MES'].tolist()
  if len(month) == 1:
    month = str(month[0])
  year = df['AÑO'].tolist()
  year = str(year[0])
  week = str(df['SEMANA'].tolist()[0])

  dic = {'Festivos': date,'AÑO':year ,'SEMANA': week,'MES': month, 'Municipio':municipios, 'Festivo':'No'}
  return dic

def agrupar_festivos(festivos, autonomia):
  list_dates = festivos["Festivos"].tolist()
  sort_dates = [(lambda x: to_date(x))(x) for x in list_dates]
  sort_dates = sorted(sort_dates)

  counter = 1
  df = pd.DataFrame()
  new_df = festivos.copy()
  for i in sort_dates:
    if counter < len(sort_dates):
      difference =  abs(i - sort_dates[counter])
      if difference == datetime.timedelta(days=2):
        date1, date2 = get_new_holidays(i, 'two')
        locate_date = str(to_string(i))

        df = festivos.loc[festivos['Festivos'] == locate_date]
        
        dic1 = create_new_row(date1, df)
        dic2 = create_new_row(date2, df)
        new_df= new_df.append(dic1, ignore_index=True)
        new_df = new_df.append(dic2, ignore_index=True)
        
      if difference == datetime.timedelta(days=1):
        date = get_new_holidays(i, 'one')
        sort_dates.append(date)

        locate_date = str(to_string(i))

        df = festivos.loc[festivos['Festivos'] == locate_date]
        dic= create_new_row(date, df)
        new_df = new_df.append(dic, ignore_index=True)
      counter += 1
  
  final_dates = [(lambda x: to_string(x))(x) for x in sort_dates]

  return new_df
```

# Focalización de filas

```python
def get_coincidentes(days, df_autonomia):
  coincident_days = []
  for i in days:
    if df_autonomia[df_autonomia['DIA']==i].size != 0:
      coincident_days.append(i)

  return coincident_days

def crear_localizador_por_fechas(festivos, df_autonomia):
  # Obtención de todos los festivos de esa comunidad
  df_festivos = pd.DataFrame()
  df_festivos['Festivos'] = festivos['Festivos'].apply(lambda x: x.strftime('%Y-%m-%d'))
  dias_festivos = (df_festivos['Festivos'].values).tolist()

  # Obtener los viajes que coinciden con dias festivos, devuelve una lista
  dias_festivos = get_coincidentes(dias_festivos, df_autonomia)

  day = dias_festivos.pop()

  df_final = df_autonomia[df_autonomia['DIA']==day]
  
  for day in dias_festivos:
    dia = df_autonomia[df_autonomia['DIA']==day]
    df_final = pd.concat((df_final, dia))

 
  return df_final
```

```python
import re
def eleccion_de_municipios(df, festivos):
  municipios = festivos['Municipio']
  
  extract = "\[|\]|\,"
  lista = []
  for i, row in festivos.iterrows():
    muni = row['Municipio']
    muni = re.sub(extract,"", muni)
    muni = muni.split("'")
    for l in muni:
      if len(l) <=2:
        pass
      else:
        lista.append(l)

  lista = list(set(lista))

  lista_uno = lista[0]
  tarjeta_final = df.loc[df['ORIGEN'] == lista_uno]

  for i in lista:
    aux_filas = df.loc[df['ORIGEN'] == i]
    tarjeta_final = pd.concat((tarjeta_final, aux_filas))

  return tarjeta_final
```
