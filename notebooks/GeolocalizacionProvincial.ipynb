En este metodo añadimos la provincia en la que empieza y acaba cada viaje así como una columna que nos dice si el viaje es interprovincial o no. Se ha elegido realizar esta operación en este momento dado que la complejidad de la operación es elevada y si la realizamos con un conjunto de datos mayor podría tardar demasiado tiempo en ejecutarse.

```python
def set_interprovincial_column(df, ccaa):
    column = []
    provincia_destino = []
    provincia_origen = []
    for i in range(len(df)):
        elemento = ccaa.loc[ccaa['Municipio'] == df.iloc[i]['ORIGEN']]['Provincia'].to_string(header=False, index=False)
        sinespacio = elemento[1:len(elemento)]
        provincia_origen.append(sinespacio)
        elemento = ccaa.loc[ccaa['Municipio'] == df.iloc[i]['DESTINO']]['Provincia'].to_string(header=False, index=False)
        sinespacio = elemento[1:len(elemento)]
        provincia_destino.append(sinespacio)

        if provincia_origen[-1] == provincia_destino[-1]:
            column.append("Si")
        else:
            column.append("No")

    df['PROVINCIA_ORIGEN'] = provincia_origen
    df['PROVINCIA_DESTINO'] = provincia_destino
    df['INTERPROVINCIAL'] = column
    return df
```

*   Andalucía
*   Castilla-La Mancha

# Localizando festivos

Guardamos en dos dataframes diferentes todos los festivos pertenecientes a los días en los que hay una festividad en Andalucía y CLM

```python
def get_fechas(ccaa):
  nacionales = df_calendario[['Fiestas nacionales']]
  espeficico = df_calendario[[str(ccaa)]]

  fiestas = nacionales.merge(nacionales)
  fiestas = fiestas.rename(columns={'Fiestas nacionales':'Festivos'})
                          
  return fiestas
```

```python
festivos_andalucia = get_fechas('Andalucía')
festivos_andalucia.tail(7)
```

```python
festivos_clm = get_fechas('Castilla-La Mancha')
festivos_clm.tail(7)
```

## Uniendo días festivos

En este paso unificamos los festivos autonómicos y por municipio, como resultado obtenemos un dataframe del tipo Festivo/Municipio, donde municipio es una lista con todos los municipios en los que ese día es fiesta. Esto es un paso intermedio para comparar una vez realizado el cluster de festivos.

```python
def unificacion_festivos(df_festivos_ccaa, df_festivos_municipios):
  municipios = list(set(df_festivos_municipios['Municipio'].to_numpy().tolist()))
  numero_elementos_columna = (df_festivos_ccaa.shape)[0]
  columna_municipios = [municipios]*numero_elementos_columna
  df_festivos_totales = df_festivos_ccaa.copy()
  df_festivos_totales['Municipio'] = columna_municipios

  df_festivos_municipios = df_festivos_municipios.rename(columns={'Fecha':'Festivos'})

  df_festivos_totales = pd.concat((df_festivos_totales, df_festivos_municipios))

  return df_festivos_totales
```

```python
df_totalFestivosAndalucia = unificacion_festivos(festivos_andalucia, df_festividadesAndalucia)
df_totalFestivosAndalucia.head(3)
```

```python
df_totalFestivosCLM = unificacion_festivos(festivos_clm, df_festividadesCLM)
df_totalFestivosCLM
```

```python
df_totalFestivosAndalucia.to_csv('df_totalFestivosAndalucia.csv')
df_totalFestivosCLM.to_csv('df_totalFestivosCLM.csv')

files.download('df_totalFestivosCLM.csv')
files.download('df_totalFestivosAndalucia.csv')
```

# Eligiendo datos

```python
def get_coincidentes(days, df_autonomia):
  coincident_days = []
  for i in days:
    if df_autonomia[df_autonomia['DIA']==i].size != 0:
      coincident_days.append(i)

  return coincident_days

def crear_localizador_por_fechas(festivos, df_autonomia):
  # Obtención de todos los festivos de esa comunidad
  dias_festivos = (festivos['Festivos'].values).tolist()

  # Obtener los viajes que coinciden con dias festivos, devuelve una lista
  dias_festivos = get_coincidentes(dias_festivos, df_autonomia)
  print(len(list(set(dias_festivos))))

  df_final = df_autonomia[df_autonomia['DIA']==dias_festivos.pop()]
  for day in dias_festivos:
    dia = df_autonomia[df_autonomia['DIA']==day]
    df_final = pd.concat((df_final, dia))

 
  return df_final

def eleccion_de_municipios(df, festivos):
  municipios = festivos['Municipio']
  lista_municipios = municipios[1].to_numpy().tolist()
  lista = lista_municipios[0]

  tarjeta_final = df[df['ORIGEN'] == lista.pop()]

  for i in lista:
    
    aux_filas = df[df['ORIGEN'] == i]

    tarjeta_final = pd.concat((tarjeta_final, aux_filas))

  return tarjeta_final
  
```

A continuación se crea una tarjeta de datos con los dias en donde es festivo de Andalucía.

```python
df_AndaluciaLocalizadoFestivos = crear_localizador_por_fechas(df_totalFestivosAndalucia, df_AndaluciaLocalizado)
df_AndaluciaLocalizadoFestivos = eleccion_de_municipios(df_AndaluciaLocalizadoFestivos, df_totalFestivosAndalucia )
print(df_AndaluciaLocalizadoFestivos.shape)
df_AndaluciaLocalizadoFestivos.head(5)
```

```python
df_CLMLocalizadoFestivos = crear_localizador_por_fechas(df_totalFestivosCLM, df_CLMLocalizado)
df_CLMLocalizadoFestivos = eleccion_de_municipios(df_CLMLocalizadoFestivos, df_totalFestivosCLM)
print(df_CLMLocalizadoFestivos.shape)
df_CLMLocalizadoFestivos.head(5)
```
